# COMBINACOES DE ID_INTERNOS

```python
def to_int(entry_int):
    try:
        aux = int(entry_int)
        return aux
    except:
        return 0
ausenc['idint'] = ausenc.loc[:,' id_interno'].apply(to_int)
formac.id_interno = formac.id_interno.apply(to_int)
```

```python
# COMBINACOES DE ID_INTERNO
carga_ids = set(carga.id_interno)
ausenc_ids = set(ausenc.loc[:,'idint'])
serv_ids = set(serv_at.id_interno)
formac_ids = set(formac.id_interno)
print('Numero de ids no cadastro de formacao: ', len(formac_ids))
print('Numero de ids no cadastro de servidores: ', len(serv_ids))
print('Numero de ids na carga horaria: ', len(carga_ids))
print('Numero de ids com ausencias', len(ausenc_ids))
print('Ids no cadastro e com ausencias: ', len(serv_ids.intersection(ausenc_ids)), ' QSE TODOS')
print('Ids no cadastro e na carga horaria: ', len(serv_ids.intersection(carga_ids)), ' TODOS')
print('Ids no cadastro e nas formacoes: ',len(serv_ids.intersection(formac_ids)),' QSE TODOS')
print('Ids na carga horaria e com ausencias: ', len(ausenc_ids.intersection(carga_ids)), ' POUCOS')
```
